## Git使用入门

### *注意事项
#### **<font color='blue'>注意事项1: token</font>** 
生成令牌: 保存在微信中
- 修改现有项目的url：git remote set-url origin  https://"your_token"@github.com/"USERNAME"/"REPO".git; 将"your_token"换成你自己得到的令牌,"USERNAME"是你自己github的用户名，"REPO"是你的项目名称.
- 然后再执行git pull，世界又恢复原样了.
- 对于全新版本，克隆的时候也在github.com前面加个令牌就好了: git clone https://"TOKEN"@github.com/"user_name"/"repo_name".git
> - [Git 使用之鉴权失败](https://blog.csdn.net/Zhao_knight/article/details/121527347)
> - [git/github采用token进行认证访问](https://www.jianshu.com/p/ec68ab0a103f)

#### **<font color='blue'>注意事项2: 更改默认初始分支</font>**
- git config --global init.defaultBranch <名称>

#### **<font color='blue'>注意事项2：将master合并入main分支</font>**

自建的github仓库存在两个分支：main和master，之前代码push到master分支，需将master分支上的代码合并到main分支。（2021年8月后，github默认分支为main）操作如下：

- git checkout main (切换到分支 'main')
- git branch （查看本地分支：包含'main'和‘master’）
- git branch -r (查看远程分支：包含'origin/main'和‘origin/master’)
- git branch --set-upstream-to=origin/main main (将分支与上游分支 'origin/main' 一致)
- git merge master --allow-unrelated-histories (将master分支合并到main上;Merge made by the 'recursive' strategy)
- git pull --allow-unrelated-histories
- git pull origin main
- git push origin main

#### **<font color='blue'>注意事项3: 解决“fatal: 远程 origin 已经存在”错误</font>**

- git remote rm origin （删除远程配置）
- git remote add origin 项目链接 （重新添加）
- git fetch origin （获取远程更新）
- git merge origin/master （把更新的内容合并到本地分支）
- git branch --set-upstream-to=origin/master master （需要将远程仓库和本地仓库关联起来）

- git pull --allow-unrelated-histories （然后使用git pull整合远程仓库和本地仓库；忽略版本不同造成的影响）


#### **<font color='blue'>注意事项4: git clone某个仓库时重定向到了另一个仓库</font>**
- **出现原因：** 在~/.gitconfig文件中设置了remote.origin.url，导致git clone时总是重定向到该url
- **解决方法：** 通过vim ~/.gitconfig打开文件，并注释掉它（或如参考资料中进行操作）
> 参考资料：[git 添加和删除 global 的 remote.origin.url](https://siwei.blog.csdn.net/article/details/79761559?spm=1001.2101.3001.6650.1&utm_medium=distribute.pc_relevant.none-task-blog-2%7Edefault%7ECTRLIST%7ERate-1-79761559-blog-105084142.pc_relevant_multi_platform_featuressortv2dupreplace&depth_1-utm_source=distribute.pc_relevant.none-task-blog-2%7Edefault%7ECTRLIST%7ERate-1-79761559-blog-105084142.pc_relevant_multi_platform_featuressortv2dupreplace&utm_relevant_index=1)

### 4、<font color='blue'>Git常用命令介绍</font>

#### 4.1 git status： 查看仓库的状态
#### 4.2 git init：初始化 Git 仓库
#### 4.3 git add：将文件添加到 Git 仓库
- git add命令并没有把文件提交到 Git 仓库，而是把文件添加到了「临时缓冲区」，这个命令有效防止了我们错误提交的可能性。
- **git rm --cached 文件** 可以取消提交

#### 4.4 git commit：将文件提交到 Git 仓库

#### 4.5 git log：打印 Git 仓库提交日志
- 提交记录的内容包括Author提交作者、Date提交日期和提交信息。

#### 4.6 git branch：查看 Git 仓库的分支情况
- 创建新分支：输入命令git branch branchName

#### 4.7 git checkout: 切换分支
- 输入git checkout branchName命令，切换到branchName分支

#### 4.8 git merge：合并分支
- 切换到master分支，然后输入git merge branchName命令，将branchName分支合并到master分支
- 在合并分支的时候，要考虑到两个分支是否有冲突，如果有冲突，则不能直接合并，需要先解决冲突；反之，则可以直接合并

#### 4.9 git branch -d & git branch -D: 删除分支
- 在命令行窗口的光标处，输入git branch -d branchName命令，删除branchName分支.
- 不过有的时候，通过git branch -d命令可以出现删除不了现象，例如分支branchName的代码没有合并到主分支等，这时如果我们一定要删除该分支，那么我们可以通过命令git branch -D进行强制删除.

#### 4.10 git tag: 为当前分支添加标签
- 输入git tag label命令，为当前分支添加标签（标签名字为 label）
- 通过命令git tag即可查看标签记录
- 通过命令git checkout label即可切换到该标签下的代码状态

### 5、<font color='blue'>利用 SSH 完成 Git 与 GitHub 的绑定</font>

#### 5.1 生成 SSH key
输入ssh-keygen -t rsa命令，表示我们指定 RSA 算法生成密钥，然后敲三次回车键，期间不需要输入密码，之后就就会生成两个文件，分别为id_rsa和id_rsa.pub，即密钥id_rsa和公钥id_rsa.pub. 对于这两个文件，其都为隐藏文件，默认生成在以下目录：
- Linux 系统：~/.ssh
- Mac 系统：~/.ssh
- Windows 系统：C:\Documents and Settings\username\\.ssh
- Windows 10 ThinkPad：C:\Users\think\.ssh

密钥和公钥生成之后，我们要做的事情就是把公钥id_rsa.pub的内容添加到 GitHub，这样我们本地的密钥id_rsa和 GitHub 上的公钥id_rsa.pub才可以进行匹配，授权成功后，就可以向 GitHub 提交代码啦！

#### 5.2 添加 SSH key
进入我们的 GitHub 主页，先点击右上角所示的倒三角▽图标，然后再点击Settins，进行设置页面；

进入Settings页面后，再点击SSH and GPG Keys进入此子界面，然后点击New SSH key按钮；

我们只需要将公钥id_rsa.pub的内容粘贴到Key处的位置（Titles的内容不填写也没事），然后点击Add SSH key 即可。

#### 5.3 验证绑定是否成功
我们可以通过在 Git Bash 中输入ssh -T git@github.com进行测试。

### 6、<font color='blue'>通过 Git 将代码提交到 GitHub</font>
#### 6.1 push命令
该单词直译过来就是“推”的意思，如果我们本地的代码有了更新，为了保持本地与远程的代码同步，我们就需要把本地的代码推到远程的仓库，代码示例：
- git push origin main

最完整的push命令是应该写成这样的：git push origin test:cz/test
- 我们注意到这里用了一个奇怪的写法test:cz/test，它的意思是说将本地的test分支推送到远程作为cz/test分支。如果我们想要本地的名称和远程一样，我们可以省略简写成：git push origin test。

如果我们设置过当前test分支的上游是远程的test，或者本地的test就是从origin拷贝过来的，那么我们可以直接git push，它会自动将本地的分支与远程关联上，会方便很多。实际上我们大多数的push操作都是这么进行的。将本地分支和远程建立映射可以使用这个命令：
- git branch --set-upstream-to master origin/master

它表示的是将本地的master和远程的master进行关联，设置过关联之后我们只需要git push和git pull就可以更新和推送这个分支了，会方便很多。

#### 6.2 pull命令
该单词直译过来就是“拉”的意思，如果我们远程仓库的代码有了更新，同样为了保持本地与远程的代码同步，我们就需要把远程的代码拉到本地，代码示例：
- git pull origin main

#### 6.3 fetch和pull
- 说到这里就需要提一下git pull和git fetch的区别了，这两者从表面上来看都是拉取远程的改动。但是两者针对的范围不同，git fetch针对远程的所有改动，而git pull只针对当前分支对应的远程分支
- 另外git pull执行之后会将远程的改动merge到本地的分支，也就是说它其实多了一步merge的操作。


> 一般情况下，我们在push操作之前都会先进行pull操作，这样不容易造成冲突。

#### 6.4 <font color='blue'>提交代码<font>
#### 6.4.1 <font color='red'>第一种提交代码方式</font>
本地没有 Git 仓库，这时我们就可以直接将远程仓库clone到本地。通过clone命令创建的本地仓库，其本身就是一个 Git 仓库了，不用我们再进行init初始化操作啦，而且自动关联远程仓库。我们只需要在这个仓库进行修改或者添加等操作，然后commit即可。
- 第一步：进入 GitHub 个人主页，点击项目，点击Clone or download，复制地址链接。
- 然后，进入我们准备存储Git仓库的目录，输入
    - git clone 地址链接
- 把需要添加的内容放到该目录中，然后输入git status命令查看仓库状态；会显示有文件未被追踪，也就是我们刚刚添加的文件。
- 在之前的文件中，我们已经知道了在真正提交代码之前，需要先进行git add操作
- 通过git commit进行提交
- 输入git log命令查看仓库日志
- 我们将本地仓库的内容push到远程仓库，输入git push origin main

#### 6.4.2 <font color='red'>第二种：提交代码方式</font>
本地有 Git 仓库，并且我们已经进行了多次commit操作。
- 第一步：首先，建立一个本地仓库，并命名
- 第二步：进入该仓库，进入init初始化操作
- 第三步：输入 git remote add origin yourLink 关联远程仓库（在此，默认大家都知道如何获取远程仓库的地址），其中origin为远程仓库的名字.
 > 可能需要执行命令：git config --global http.sslVerify "false"
- 第四步：输入git pull origin main命令，同步远程仓库和本地仓库
- 第五步：接下来，在本地仓库新建文件
- 第六步：输入git add和git commit命令，将文件添加并提交到仓库
- **重要：** 还需要如下命令：
 > git remote set-url origin https://"token"@github.com/"userName"/"repo".git
- 第七步：再输入git push origin main命令，将本地仓库修改（或者添加）的内容提交到远程仓库


#### *提示

>- 一般来说，我们习惯性将远程仓库命名为origin，不过在需要关联多个远程仓库的时候，就需要我们再取别的名字啦！
>-  再强调一遍：在我们向远程仓库提交代码的时候，一定要先进行pull操作，再进行push操作，防止本地仓库与远程仓库不同步导致冲突的问题，尤其是第二种提交代码的情况，很容易就出现问题。

#### 6.4.3 <font color=red>修改文件后再提交</font>
修改文件后，怎么提交到远程仓库
- git status 查看git是否有修改内容需要提交
- git add 指向需要提交的内容文件
- git commit 提交到本地库
- 执行如下两个命令(**视情况看是否需要执行这一条**)：
